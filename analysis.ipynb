```python
# importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# import library for cloud of words
# from wordcloud import WordCloud, STOPWORDS
```

```python
# reading the dataset
df = pd.read_csv('news.csv')
```

```python
# performing some basic analysis
print(df.head())
print(df.info())
print(df.describe())
```

```python
# rename the category names 0-7 to the actual category names
df['Category'] = df['Category'].replace(1, 'Sports')
df['Category'] = df['Category'].replace(2, 'Politics')
df['Category'] = df['Category'].replace(3, 'TV Showbiz')
df['Category'] = df['Category'].replace(4, 'International News')
df['Category'] = df['Category'].replace(5, 'Technology')
df['Category'] = df['Category'].replace(6, 'Business')
df['Category'] = df['Category'].replace(7, 'Health')
```

```python
# plot some graphs
# plot a graph for the number of news in each category
# Rachel Daly With A Nice Curler,1

# plot a graph for the number of news in each category
# label the x-axis as news and y-axis as category in the dataset
plt.figure(figsize=(12, 8))
sns.countplot(x='Category', data=df)
plt.xlabel('Title')  
plt.ylabel('Category')
plt.show()
```

```python
# provide a key of what each category means
print(df['Category'].value_counts())
```

```python
print("1: Sports")
print("2: Politics")
print("3: TV Showbiz")
print("4: International News")
print("5: Technology")
print("6: Business")
print("7: Health")
```

```python
# pie chart
# plot a pie chart for the number of news in each category
# label the x-axis as news and y-axis as category in the dataset

plt.figure(figsize=(12, 8))
df['Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```
